Int, or integer, is a whole number, positive or negative, without decimals, of unlimited length.

Float, or "floating point number" is a number, positive or negative, containing one or more decimals.

Strings in python are surrounded by either single quotation marks, or double quotation marks.

```python
a = 1    # int
b = 2.8  # float

x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print(x)
print(y)
print(z)

print("Hello") # display a string literal with the print() function
c = "Hello"
print(c)
```

Strings cannot be modified, but Lists can be modified. Lists are used to store multiple items in a single variable. The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.

```python
thislist = ["apple", "banana", "cherry"]
print(thislist)
# change 2nd item
thislist[1] = "blackcurrant"
print(thislist)
# Insert "watermelon" as the third item
thislist.insert(2, "watermelon")
print(thislist)
# add an item to the end of the list
thislist.append("orange")
print(thislist)
# Insert an item as the second position
thislist.insert(1, "orange")
print(thislist)
```
